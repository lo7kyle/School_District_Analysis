```python
# separate the student names from main df so we wont affect main df
# use tolist() to all student names to put names in list.
student_names = student_data_df["student_name"].tolist()
student_names
```

```python
# we want to separate names with 3 or more len and put them in diff list to do more analysis. use conditonals
#create new list and use it for for-loop to iterate through list
students_to_fix =[]
#use if statement to check the length of name and if >= to 3, if true add to students_to_fix
for name in student_names:
        if len(name.split()) >= 3: 
            students_to_fix.append(name)
# find how many students names we need to fix
len(students_to_fix)
```

```python
# prefixes we want to remove:
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
#use python .str method to turn object into str: ie. student_data_df["student_name"].str.replace()
#Iterate through the prefixes_suffixes list and replace with "" when it appears in student_data_df
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
student_data_df.head(10)
```

```python
# use merge() method to merge 2 df with similar column names
# we will add df and shared columns to each df and define l,r,inner, or outer
#combining data into single data set
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# use count to check total number of students were corrected and not dropped
student_count = school_data_complete_df.count()
student_count
```

```python
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
#we want the count of schools, so assign school_data_df a variable to one of its columns
school_count = school_data_df["school_name"].count()
school_count
```

```python
#if we want to use complete df, we need to find unique values first which gives a list, then we take len of list to get total 
# count
school_count_2 = len(school_data_complete_df["school_name"].unique())
school_count_2
```

```python
#find the sum of the budget using sum() on school_data_df
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#using mean() method we can get average for reading/math scores
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
#average math score
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#we want to get the number of students who passed math and reading. 
#to do that we need to create 2 variables for both math and reading and use >=70 to determine passing
# we can filter the data by passing the dataframe with ["math_score"] within itself
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >=70]
passing_math.head()
```

```python
#do the same with passing_reading
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
#now we calculate how many students passed by using count() method
passing_math_count = passing_math["student_name"].count()

#count number of student who passed reading
passing_reading_count = passing_reading["student_name"].count()

print(passing_math_count)
print(passing_reading_count)
```

```python
# get the percentage by dividing count of pass by total number of students * 100
student_count=school_data_complete_df["Student ID"].count()

passing_math_percentage = (passing_math_count/ float(student_count)) * 100
passing_reading_percentage = (passing_reading_count/ float(student_count)) * 100

print(passing_math_percentage)
print(passing_reading_percentage)
```

```python
# we want overall both math and reading combine. use similar method above, but use & conditional
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"]>=70) & (school_data_complete_df["reading_score"]>=70)]

passing_math_reading.head()
```

```python
# now get the count and percentage of combined / total students * 100
overall_math_reading_count = passing_math_reading["student_name"].count()
overall_math_reading_count
```

```python
# percentage of overall
overall_passing_percentage = overall_math_reading_count/ (float(student_count)) * 100
overall_passing_percentage
```

```python
# with the analysis done, we can now add the data into a new df
# recall we can do this by converting a list of dictionaries into a data frame
district_summary_df = pd.DataFrame(
    [{"Total Schools": school_count,
     "Total Students": student_count,
     "Total Budget": total_budget,
     "Average Math Score": average_math_score,
     "Average Reading Score": average_reading_score,
     "% Passing Math": passing_math_percentage,
     "% Passing Reading": passing_reading_percentage,
     "% Overall Passing": overall_passing_percentage}]
    )
district_summary_df
```

```python
# we want to add dollar sign to budget and round average score to only tenths place.
# use map() function to substitue value in a series with another value
# Format "Total Students" to have a comma for thousands seperator
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Students"]
```

```python
# now we want to format the budget and add a , and $ sign and have 2 decimal places
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Budget"]
```

```python
# formatting the columns 
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
```

```python
# formatted district_summary_df 
district_summary_df
```

```python
# we can reorder our df using square bracket notation
#syntax: new_column_order = ["column 2", "column 4", "column 1"]
#assign new or the same df to the new column order
# df = df[new_column_order]
# Reorder the columns in the order you want them to appear.
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]

# Assign district summary df the new column order.
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```

```python
# we want to find average math and reading scores. use mean() method to get average and groupby() to split an object like a df
# and apply a mathematical operation because we cant do arithmetic on an object
#first find math scores of each student at each school using students_complete data
student_school_math = student_data_df.set_index(["school_name"])["math_score"]
student_school_math
```

```python
# now that we have all of the students scores, we want to find mean(). use groupby to perform mathematical operations
per_school_averages = school_data_complete_df.groupby(["school_name"]).mean()
per_school_averages
```

```python
# average test score of math and reading
per_school_math = school_data_complete_df.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]

print(per_school_math)
print(per_school_reading)
```

```python
# now we want passing percentages. using the old code from earlier
# Calculate the passing scores by creating a filtered DataFrame.
per_school_passing_math = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)]

per_school_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)]

per_school_passing_math
```

```python
# we want to group by schools and use count to get student count who passed math/reading
per_school_passing_math = per_school_passing_math.groupby(["school_name"]).count()["student_name"]
per_school_passing_reading = per_school_passing_reading.groupby(["school_name"]).count()["student_name"]

#print out to check if we got what we want
print(per_school_passing_math)
print(per_school_passing_reading)
```

```python
# now we can calculate the percentage of students passing math and reading by dividing per_school passing with total count 
per_school_passing_math = per_school_passing_math / per_school_count * 100

per_school_passing_reading = per_school_passing_reading / per_school_count * 100

per_school_passing_math
```

```python
# now calculate passing both math/reading using old code with &
per_passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
per_passing_math_reading.head()
```

```python
per_passing_math_reading = per_passing_math_reading.groupby(["school_name"]).count()["student_name"]

per_passing_math_reading
```

```python
# calculate percentage
per_overall_passing_percentage = per_passing_math_reading / per_school_count * 100

per_overall_passing_percentage
```

```python
# create list of dictionaries with values we want in column order we want
per_school_summary_df = pd.DataFrame({
    "School Type": per_school_types,
    "Total Students": per_school_count,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": per_overall_passing_percentage
})
per_school_summary_df.head()
```
