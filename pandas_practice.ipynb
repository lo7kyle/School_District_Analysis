```python
# A dictionary of high schools and the type of school.
high_school_types = [{"High School": "Griffin", "Type":"District"},
                    {"High School": "Figueroa", "Type": "District"},
                    {"High School": "Wilson", "Type": "Charter"},
                    {"High School": "Wright", "Type": "Charter"}]
```

```python
#skill practice iterating thorugh list of dictionary
#This one iterates and prints out the keys and the values of the dictionaries in the list
for index in range(len(high_school_types)):
    for key in high_school_types[index]:
        print(high_school_types[index][key])
```

```python
#Another way to print out the list of dictionary
#This way only prints out the dictionaires in the list
for l_dict in high_school_types:
    print (l_dict)
    
```

```python
# A data frame consists of more than 1 column and is 2-Dimensional
#we can convert a list of dictionaries into a dataframe
# A dictionary of high schools
high_school_dicts = [{"School ID": 0, "school_name": "Huang High    School", "type": "District"},
                   {"School ID": 1, "school_name": "Figueroa High School", "type": "District"},
                    {"School ID": 2, "school_name":"Shelton High School", "type": "Charter"},
                    {"School ID": 3, "school_name":"Hernandez High School", "type": "District"},
                    {"School ID": 4, "school_name":"Griffin High School", "type": "Charter"}]
```

```python
#using pandas to convert list of dicts into dataframe
school_df = pd.DataFrame(high_school_dicts)
school_df
```

```python
#converting a list or series to a dataframe 
# we can create the same df by adding each column as a list or series to an empty df
# Three separate lists of information on high schools
#column for ids
school_id = [0, 1, 2, 3, 4]
#column for school names
school_name = ["Huang High School", "Figueroa High School",
"Shelton High School", "Hernandez High School","Griffin High School"]
#column for school type
type_of_school = ["District", "District", "Charter", "District","Charter"]
```

```python
#We can initialize an empty df
schools_df = pd.DataFrame()
```

```python
# we can add the key to a dictionary by putting single/double quotes inside a bracket
# counties_dict["Arapahoe"] = 422829
#adding the list to a new df
schools_df["School ID"] = school_id
#print the df
schools_df
```

```python
#adding list school_name, type_of_school to df
schools_df["school_name"] = school_name
schools_df["type"] = type_of_school
#print df
schools_df
```

```python
# another way to create this data frame is to create a python dicionary from 3 lists
# Create a dictionary of information on high schools.
high_schools_dict = {'School ID': school_id, 'school_name':school_name, 'type':type_of_school}
high_schools_dict
```

```python
# check to see if you can convert dicionary of lists into df
list_of_schools = pd.DataFrame(high_schools_dict)
list_of_schools
```

```python
# add to the df
# we can either add to the list already created by using append
# or we can create a new list and add to the df

#This method we will create another list, then add to the current list
school_id_add = range(5,15)
#column for school names
school_name_add = ("Wilson High School", "Cabrera High School", "Bailey High School",
                   "Holden High School","Pena High School", "Wright High School",
                   "Rodriguez High School","Johnson High School","Ford High School",
                   "Thomas High School")
#column for school type
type_of_school_add = ("Charter","Charter","District",
                     "Charter","Charter","Charter",
                     "District","District","District",
                     "Charter")
# now add to original list and print out the new list
school_id.extend(school_id_add)
school_name.extend(school_name_add)
type_of_school.extend(type_of_school_add)
print(school_id)
print(school_name)
print(type_of_school)
```

```python
#Create a dataframe by creating a dictionary from the lists
high_schools_dict = {'School ID': school_id, 'school_name':school_name, 'type':type_of_school}
dict_df = pd.DataFrame(high_schools_dict)
dict_df
```

```python
#you can also use append in pandas df.append()
new_high_schools_dict = {'School ID': school_id_add, 'school_name': school_name_add, 'type': type_of_school_add}
new_school_df = pd.DataFrame(new_high_schools_dict)
new_school_df
```

```python
#appending new_school_df to school_df. the ignore_index = True is for 
schools_appended = schools_df.append(new_school_df, ignore_index = True)
schools_appended
```
